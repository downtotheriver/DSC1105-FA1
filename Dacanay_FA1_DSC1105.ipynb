## Importing libraries

```r
library(readr)
library(crayon)
library(ggplot2)
```

## **Data**

```r
cytof <- read_csv("/kaggle/input/cytof-one-experiment/cytof_one_experiment.csv", show_col_types = FALSE)
cytof
```

## I. Analyzing CD4
### A. Range of 'CD4' columm

```r
cat(bold("Min:"), min(cytof$CD4), "\n")
cat(bold("Max:"), max(cytof$CD4))
```

### B. Quantile and ECDF of CD4 Expression Level

```r
cat(bold("QUANTILE"), "\n")
quantile(cytof$CD4)
cat("\n")

ggplot(cytof, aes(x = CD4)) +
    stat_ecdf(geom = "step", color = "blue", linewidth = 0.5) +
    labs(title = "ECDF of CD4 Expression Level",
        x = "Expression Level",
        y = "ECDF") +
        theme_minimal() +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

As shown above, our **CD4 expression levels ranges from -0.673865 to 4.932248**. Moreover, approximately **50%** of the CD4 expression levels **fall below -0.2**. This means that half the cells have **very low or no CD4 expression**. Hence only a **minority of cells are CD4+ helper T cells**. On the other hand, this implies that the **remaining half** of the samples approximately **have CD4 expression levels between -0.2 and 5**.

Furthermore, the figure above displays a steep curve approximately for **values between -0.6 and 0**, suggesting that **plenty of samples are within that CD4 expression levels**. For the remaining samples, the expression levels are somewhat evenly distributed.

### C. Density Estimate

```r
ggplot(cytof, aes(x = CD4)) +
    geom_density(fill = "darkblue", alpha = 0.5) +
    labs(title = "Density Estimate of CD4 Expression Level",
        x = "Expression Level", y = "Density") +
    theme_minimal() +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

The figure above illustrates the CD4 expression level distribution. As shown above, a tall curve covers the values from about -0.6 to 0. This implies that **majority of the samples fall below the expression level of 0**.

## II. Comparison of CD4 and CD8
### A. Q-Q Plot of CD4 vs CD8 Expression

```r
colnames(cytof)
```

```r
CD4 <- sort(cytof$CD4)
CD8 <- sort(cytof$CD8)

n <- min(length(CD4), length(CD8))
qq_data <- data.frame(
    col1 = CD4[1:n],
    col2 = CD8[1:n]
)

ggplot(qq_data, aes(x = col1, y = col2)) +
    geom_point(color = "blue") +
    labs(title = "Q-Q Plot: CD4 vs CD8",
        x = "CD4",
        y = "CD8") +
    theme_minimal() +
    theme(plot.title = element_text(hjust = 0.5, size = 20))
```

In the lower left part of the graph, the line deviates significantly above the diagonal line, indicateing that **the lower quantiles of the CD8 data are higher than the corresponding lower quantiles of the CD4 data**. This means that there maybe more low values in CD4 compared to CD8, or the lowest values in CD4 are notably smaller than the lowest values in CD8.
